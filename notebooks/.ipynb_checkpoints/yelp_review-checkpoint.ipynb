### 3 .Analisis of Review dataset

First we are loading the libraries and read the dataset. After all intentions to read this size dataset, we ended up dividing data into chunks and collecting.

```python
import pandas as pd
import json
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#we define a set for divided chunks
chunks = []

#creating jsonreader object of size 1000 to be collected again.
reader = pd.read_json('../data/raw/yelp_academic_dataset_review.json', 
                      chunksize = 1000,
                     lines = True)
```

```python
#loop to collect all objects in chunks
for chunk in reader:
    
    chunks.append(chunk)
    
#merging chunks to dataframe    
df_reviews = pd.concat(chunks)
                          
```

```python

print("Dataset has ", len(df_reviews.index), "rows and ", len(df_reviews.columns), "columns")
df_reviews.head()
```

```python
# We create a small df to understand feature types and null values.
reviews_dtypes = pd.merge(df_reviews.isnull().sum(axis = 0).sort_values().to_frame('missing_value').reset_index(),
         df_reviews.dtypes.to_frame('feature_type').reset_index(),
         on = 'index',
         how = 'inner')
reviews_dtypes
```

In this dataset there isn't any null values. 

review_id, user_id and business_id is not useful for a machine learning model. But in our case we keep it to connet with other datasets.

Other things we see in the first glimpse, variables stars, useful, funny and cool are numeric and are integer. In business dataset, since the star value vas representing mean of the all reviews, they were in float.

useful, funny, cool seems to have binary values, if reader of the review find it useful or no. We can check if this is help us to explain. 

text is a text value. Although this is not numeric we all know that we can infere things from reviews with the help of some libraries in python. Therefore, it can be an important variable in our analisis.



```python
#We create a groupby df to see the distribution of stars.

df_review_count = df_reviews.groupby(by=['stars'], dropna = False).count().reset_index()
df_review_count
```

```python
fig = plt.figure(figsize=(8, 6))
sns.barplot(x = 'stars', y = 'review_id', data = df_review_count)
plt.title("Review count vs Rating",fontsize = 20)
plt.ylabel('Number of Reviews', fontsize = 12)
plt.xlabel('Rating', fontsize = 12)
```

Around half of the reviews apear as 5. Users rate good qualifications more than bad qualifications.

```python
#Now we do the same for usefulness of the reviews.

df_review_useful = df_reviews.groupby(by=['cool'], dropna = False).count().reset_index()
df_review_useful
```
