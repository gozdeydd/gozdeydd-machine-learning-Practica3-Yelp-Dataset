### Autoencoders
Vamos a utilizar una arquitectura de autoencoders para poder entrenar un modelo de Deep Learning en un dataset 'pequeño'.

```python
import matplotlib.pyplot as plt
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras import layers, Input, regularizers
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import SGD, Adam
from tensorflow.keras.metrics import AUC
from tensorflow.keras.losses import SparseCategoricalCrossentropy, MSE, MAE
import tensorflow as tf
from aux_func import show_history, r2_keras
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Dense, Flatten, BatchNormalization, Conv2DTranspose
from tensorflow.keras.callbacks import EarlyStopping
from time import time
from tensorflow.keras.layers import Dropout
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2
from tensorflow.data import Dataset
import time
```

Utilizaremos el dataset CIFAR 10

```python
from tensorflow.keras.datasets import cifar10

(X_train_cifar10, y_train_cifar10), (X_validation_cifar10, y_validation_cifar10) = cifar10.load_data()
X_train_cifar10 = X_train_cifar10/255
X_validation_cifar10 = X_validation_cifar10/255

labels = ['airplane', 'automobile', 'bird', 'cat', 'deer', 'dog', 'frog', 'horse', 'ship', 'truck']

print('Número de elementos en el dataset = ',X_train_cifar10.shape[0]+X_validation_cifar10.shape[0])
plt.title(labels[int(y_train_cifar10[62])])
plt.imshow(X_train_cifar10[62])
plt.show()
```

Lo primero que vamos a hacer es entrenar una red convolucional estándar en un subconjunto pequeño (300 elementos) del dataset CIFAR.
Y veremos que el score obtenido no es bueno.

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau
# capas de la red
input = Input(shape=(32,32,3))
layer = input
layer = Conv2D(filters=25, kernel_size=3, strides=1, padding='same', activation='relu')(layer)
layer = MaxPooling2D((2, 2))(layer)
layer = BatchNormalization()(layer)
layer = Conv2D(filters=50, kernel_size=3, strides=1, padding='same', activation='relu')(layer)
layer = MaxPooling2D((2, 2))(layer)
layer = BatchNormalization()(layer)
layer = Conv2D(filters=100, kernel_size=3, strides=1, padding='same', activation='relu')(layer)
layer = MaxPooling2D((2, 2))(layer)
layer = BatchNormalization()(layer)
layer = Flatten()(layer)
layer = Dropout(0.7)(layer)
layer = Dense(units=100, activation='relu')(layer)
output = Dense(units=10, activation='softmax')(layer)

# creamos el modelo
model = Model(inputs=input, outputs=output)
print(model.summary())

# optimizador
optimizer = Adam(learning_rate=0.001)

# función loss
loss = SparseCategoricalCrossentropy(from_logits=False)

# métrica
metrics = ['accuracy']

# compilamos el modelo
model.compile(optimizer=optimizer, loss=loss, metrics=metrics)

# Podemos decidir que score queremos monitorizar, cuantas epochs esperar después del mejor score y si queremos que nos devuelva el mejor modelo.
reduce_lr = ReduceLROnPlateau(monitor='val_accuracy', factor=0.5, patience=5, mode='max')

history = model.fit(X_train_cifar10[:300], y_train_cifar10[:300], batch_size=30, epochs=30,
                    validation_data=(X_validation_cifar10, y_validation_cifar10))
```

```python
show_history(history)
```

El accuracy obtenido es del 14%, nada deseable comparado con un 84% de modelos anteriores.
Esto se debe a que la red no puede generalizar y extraer los patrones complejos de este dataset tan reducido.
Veamos como podemos usar un autoencoder para solventar este problema.
Vamos a suponer que disponemos solo de 300 etiquetas del dataset pero que disponemos de todas las imágenes de train.
En general es mucho más fácil disponer de los datos que de las etiquetas.

Lo primero es entrenar una red autoencoder sobre el dataset X_train_cifar10. Con ello obtenemos un codificador que es capaz de extraer características representativas del dataset.
Utilizaremos en la ultima capa una sigmoid y como función loss la mse

```python
from tensorflow.keras.callbacks import ReduceLROnPlateau
# capas de la red
input = Input(shape=(32,32,3))
layer = input
layer = Conv2D(filters=50, kernel_size=3, strides=1, padding='same', activation='relu')(layer)
layer = MaxPooling2D((2, 2))(layer)
layer = BatchNormalization()(layer)
layer = Conv2D(filters=100, kernel_size=3, strides=1, padding='same', activation='relu')(layer)
layer = MaxPooling2D((2, 2))(layer)
layer = BatchNormalization()(layer)
layer = Conv2D(filters=200, kernel_size=3, strides=1, padding='same', activation='relu')(layer)
layer = MaxPooling2D((2, 2))(layer)
layer = BatchNormalization()(layer)
layer = Conv2D(filters=400, kernel_size=3, strides=1, padding='same', activation='relu')(layer)
layer = MaxPooling2D((2, 2))(layer)
layer = BatchNormalization()(layer)
layer = Conv2D(filters=400, kernel_size=3, strides=1, padding='same', activation='relu')(layer)
layer = BatchNormalization()(layer)
layer = Conv2DTranspose(filters=400, kernel_size=3, strides=2, padding='same', activation='relu')(layer)
layer = BatchNormalization()(layer)
layer = Conv2DTranspose(filters=200, kernel_size=3, strides=2, padding='same', activation='relu')(layer)
layer = BatchNormalization()(layer)
layer = Conv2DTranspose(filters=100, kernel_size=3, strides=2, padding='same', activation='relu')(layer)
layer = BatchNormalization()(layer)
layer = Conv2DTranspose(filters=50, kernel_size=3, strides=2, padding='same', activation='relu')(layer)
output = Conv2D(filters=3, kernel_size=1, strides=1, padding='same', activation='sigmoid')(layer)

# creamos el modelo
model = Model(inputs=input, outputs=output)
print(model.summary())

# optimizador
optimizer = Adam(learning_rate=0.001)

# compilamos el modelo
model.compile(optimizer=optimizer, loss='mse', metrics=['mse', r2_keras])

history = model.fit(X_train_cifar10, X_train_cifar10, batch_size=50, epochs=20,
                    validation_data=(X_validation_cifar10, X_validation_cifar10))
```

```python
show_history(history, metrics=['mse', 'r2_keras'])
```

Hemos obtenido un modelo capaz de codificar y descodificar las imágenes. Además vemos como no se produce nada de overfiting.

Aplicamos la técnica de Transfer Learning para entrenar un nuevo modelo utilizando el codificador del autoencoder

```python
# El input del nuevo modelo será el mismo que el modelo base
input = model.input

# Seleccionamos hasta que capa del modelo base queremos conservar
last_layer_model_base = model.layers[14].output

# Creamos el modelo de clasificación que irá encima del modelo base modificado
layer = Flatten()(last_layer_model_base)
layer = Dropout(0.7)(layer)
layer = Dense(units=100, activation='relu')(layer)
output = Dense(units=10, activation='softmax')(layer)

# creamos el modelo
new_model = Model(inputs=input, outputs=output)
print(new_model.summary())

# congelamos las capas no densas para que no sean modificadas durante el entrenamiento
for layer in new_model.layers:
    if not 'dense' in layer.name:
        layer.trainable = False

# compilamos el modelo y entrenamos
new_model.compile(optimizer=optimizer, loss=loss, metrics=metrics)
history = new_model.fit(X_train_cifar10[:300], y_train_cifar10[:300], batch_size=50, epochs=30,
                    validation_data=(X_validation_cifar10, y_validation_cifar10))
```

```python
show_history(history)
```

Obtenemos un accuracy del 32%, nada mal para un dataset de 300 elementos.

Sería recomendable probar que accuracy obtenemos en función de la profundidad que seleccionamos del codificador.

### GAN
Las redes GAN son usadas para la generación sintética de datos, con ellas podemos hacer que el dataset generado sintéticamente tenga propiedades similares a las de un dataset dado.
Para ello debemos contruir un generador y un discriminador. Los cuales competirán para engañarse mutuamente.
Primero vamos a construir un generador, el cual a partir de ruido generará las imágenes. Para ello utilizaremos capas Conv2DTranspose.

```python
BUFFER_SIZE = 60000
BATCH_SIZE = 256

input = Input(shape=(100))
layer = input
layer = Dense(units=7*7*256,use_bias=False, activation=tf.nn.leaky_relu)(layer)
layer = BatchNormalization()(layer)
layer = tf.reshape(layer,(-1, 7, 7, 256))
layer = Conv2DTranspose(128, (5, 5), strides=(1, 1), padding='same', use_bias=False, activation=tf.nn.leaky_relu)(layer)
layer = BatchNormalization()(layer)
layer = Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation=tf.nn.leaky_relu)(layer)
layer = BatchNormalization()(layer)
output = Conv2DTranspose(1, (5, 5), strides=(2, 2), padding='same', use_bias=False, activation='tanh')(layer)

# creamos el modelo
generator = Model(inputs=input, outputs=output)

noise = tf.random.normal([1, 100])
generated_image = generator(noise, training=False)

plt.imshow(generated_image[0, :, :, 0], cmap='gray')
#28X28 boyutunda resim uretiyor. nerden anladik 7*strides
```

Inicialmente la red genera datos aleatorios.
Ahora vamos a construir el discriminador, el cual en función de las características de la imagen determinará si es una imagen real o sintética.
Para ello utilizaremos una red convolucional estándar.

```python
input = Input(shape=(28,28, 1))
layer = input
layer = Conv2D(filters=25, kernel_size=3, strides=1, padding='same', activation='relu')(layer)
layer = MaxPooling2D((2, 2))(layer)
layer = BatchNormalization()(layer)
layer = Conv2D(filters=50, kernel_size=3, strides=1, padding='same', activation='relu')(layer)
layer = MaxPooling2D((2, 2))(layer)
layer = BatchNormalization()(layer)
layer = Conv2D(filters=100, kernel_size=3, strides=1, padding='same', activation='relu')(layer)
layer = MaxPooling2D((2, 2))(layer)
layer = BatchNormalization()(layer)
layer = Flatten()(layer)
layer = Dropout(0.3)(layer)
layer = Dense(units=100, activation='relu')(layer)
output = Dense(units=1, activation='sigmoid')(layer)

# creamos el modelo
discriminator = Model(inputs=input, outputs=output)

decision = discriminator(generated_image)
print (decision)
```

Inicialmente la red clasifica de manera aleatoria las imágenes.

```python
cross_entropy = tf.keras.losses.BinaryCrossentropy(from_logits=False)
```

Este punto es crucial, en el definimos las loss del discriminador y generador.
Las imágenes reales las etiquetamos con 1's.

```python
def discriminator_loss(real_output, fake_output):
    # El discriminador tiene que aprender a distinguir y etiquetar correctamente las imágenes
    real_loss = cross_entropy(tf.ones_like(real_output), real_output)
    fake_loss = cross_entropy(tf.zeros_like(fake_output), fake_output)
    total_loss = real_loss + fake_loss
    return total_loss

def generator_loss(fake_output):
    # El generador tiene que aprender a engañar al discriminador
    return cross_entropy(tf.ones_like(fake_output), fake_output)

generator_optimizer = tf.keras.optimizers.Adam(1e-4)
discriminator_optimizer = tf.keras.optimizers.Adam(1e-4)
```

Parámetros de entrenamiento

```python
EPOCHS = 50
noise_dim = 100
num_examples_to_generate = 16

seed = tf.random.normal([num_examples_to_generate, noise_dim])
```

Ahora definimos el step de entrenamiento custom que vamos a necesitar para entrenar la red.

```python
# Notice the use of `tf.function`
# This annotation causes the function to be "compiled".
@tf.function
def train_step(images):
    # Generamos el ruido aleatorio para el generador
    noise = tf.random.normal([BATCH_SIZE, noise_dim])

    with tf.GradientTape() as gen_tape, tf.GradientTape() as disc_tape:
        # Creamos con el generador un batch de imágenes sintéticas
        generated_images = generator(noise, training=True)

        # Clasificamos con el discriminador las imágenes reales y sintéticas
        real_output = discriminator(images, training=True)
        fake_output = discriminator(generated_images, training=True)

        # Obtenemos la loss de ambas redes
        gen_loss = generator_loss(fake_output)
        disc_loss = discriminator_loss(real_output, fake_output)

    # Calculamos los gradientes en funciones de las losses
    gradients_of_generator = gen_tape.gradient(gen_loss, generator.trainable_variables)
    gradients_of_discriminator = disc_tape.gradient(disc_loss, discriminator.trainable_variables)

    # Optimizamos los pesos de la red utilizando el optimizador
    generator_optimizer.apply_gradients(zip(gradients_of_generator, generator.trainable_variables))
    discriminator_optimizer.apply_gradients(zip(gradients_of_discriminator, discriminator.trainable_variables))
```

```python
# Función para plotear los resultados mientras entrenamos
def generate_and_save_images(model, epoch, test_input):
    # Notice `training` is set to False.
    # This is so all layers run in inference mode (batchnorm).
    predictions = model(test_input, training=False)

    fig = plt.figure(figsize=(4, 4))

    for i in range(predictions.shape[0]):
        plt.subplot(4, 4, i+1)
        plt.imshow(predictions[i, :, :, 0] * 127.5 + 127.5, cmap='gray')
        plt.axis('off')

    #plt.savefig('image_at_epoch_{:04d}.png'.format(epoch))
    plt.show()
```

```python
# Función que itera sobre el dataset, entrena y genera imágenes para su visualización
def train(dataset, epochs):
    for epoch in range(epochs):
        start = time.time()

        for image_batch in dataset:
            train_step(image_batch)

        # Produce images for the GIF as you go
        generate_and_save_images(generator,epoch + 1,seed)


        print ('Time for epoch {} is {} sec'.format(epoch + 1, time.time()-start))

    # Generate after the final epoch
    generate_and_save_images(generator,epochs,seed)
```

```python
# Generamos y normalizamos el dataset.
(X_train_mnist, y_train_mnist), (X_validation_mnist, y_validation_mnist) = mnist.load_data()

X_train_mnist = (X_train_mnist - 127.5) / 127.5

# Creamos un Dataset utilizando numpys
train_ds = Dataset.from_tensor_slices((X_train_mnist, y_train_mnist))

# Aleatorizamos el dataset usando un buffer de 5000
train_ds = train_ds.shuffle(BUFFER_SIZE)

# Hacemos que el dataset este formado por batches
train_ds = train_ds.batch(BATCH_SIZE)
```

Entrenamos y ploteamos los números que se generan durante el entrenamiento

```python
train(train_ds, EPOCHS)
```
